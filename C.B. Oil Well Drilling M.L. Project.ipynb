```python
#Calculate the volume of reserves per well required to develop a well without having losses.


#The budget for development of 200 oil wells is 100 million USD.

well_quantity = 200

budget_USD = 10e7

#The revenue from one unit of product is 4500 USD (volume  of reserves is expressed in thousands of barrels)

revenue_USD_per_unit = 4.5*10e2

break_even_value = budget_USD/well_quantity 


#500,000/4500 = 111.111 units of product per well required to break even.

minimum_product_to_profit = break_even_value/revenue_USD_per_unit 
 

minimum_for_profit = f"{minimum_product_to_profit:.2f}"

minimum_for_profit
```

```python
def compare_breakeven_with_avg_reserves(avg_reserves_region, break_even_point, region):
    if avg_reserves_region < break_even_point:
        difference = break_even_point - avg_reserves_region
        
        percent = (difference/avg_reserves_region) * 100
       
        print(f"The average reserves of {region} are {percent} percent lower than the break even point")
        
    else:
        difference = avg_reserves_region - break_even_point
        
        percent = (difference/avg_reserves_region) *100
       
        print(f"The average reserves of {region} are {percent} percent higher than the break even point")
```

```python
def calculate_profit(predictions, n_top_wells=200, price_per_barrel=4500, budget=100e6):
    # Sort the predictions DataFrame by predicted product in descending order
#sample_wells = predictions.sample(500, random_state = 12345)#I had the wrong idea here
    
    predictions_sorted = predictions.sort_values(by='Actual', ascending=False) ##changed the by = argument to Actual
    
    # Select the top n wells
    top_wells = predictions_sorted.head(n_top_wells)
    
    # Calculate the total revenue
    total_revenue = top_wells['Actual'].sum() * price_per_barrel ##Changed from Predicted to Actual
    
    # Calculate the profit (revenue minus development costs)
    profit = total_revenue - budget
    
    return profit

#Changed first line to include only sample of 500 points, It increased the risk amount and lowered profit
```

```python
def calculate_final_profit(predictions, n_top_wells=200, price_per_barrel=4500, budget=100e6):
    # Sort the predictions DataFrame by predicted product in descending order
#sample_wells = predictions.sample(500, random_state = 12345)#I had the wrong idea here
    
    predictions_sorted = predictions.sort_values(ascending=False)
    
    # Select the top n wells
    top_wells = predictions_sorted.head(n_top_wells)
    
    # Calculate the total revenue
    total_revenue = top_wells.sum() * price_per_barrel ##Changed from Predicted to Actual
    
    # Calculate the profit (revenue minus development costs)
    profit = total_revenue - budget
    
    return profit
```
